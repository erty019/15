Задача: Вычисление элементов произведения  C  двух квадратных матриц  A=(aij)  и  B=(bij)  по формулам
cij=∑k=1naikbkj(i=1,2,…,n; j=1,2,…,n) 
легко реализуется с помощью трех вложенных циклов: по  i , по  j  и по  k . Различные варианты алгоритма можно получить перестановкой циклов. Напишите шесть функций multijk, multikj, multjik, ..., реализующих каждый из этих вариантов. Для серии случайно сгенерированных матриц больших размеров замерьте время, требуемое этими функциями и стандартной операцией @. Нарисуйте графики зависимости времени от порядка матриц. Объясните, почему требуется различное время на выполнение каждой из написанных вами функций.

```python
import time
import numpy as np
import matplotlib.pyplot as plt
#функции вычисления элементов матрицы
def multijk(A, B):
    n = len(A)
    C = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

def multikj(A, B):
    n = len(A)
    C = np.zeros((n, n))
    for i in range(n):
        for k in range(n):
            for j in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

def multjik(A, B):
    n = len(A)
    C = np.zeros((n, n))
    for j in range(n):
        for i in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
def multkij(A, B):
    n = len(A)
    C = np.zeros((n, n))
    for k in range(n):
        for i in range(n):
            for j in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

def multjki(A, B):
    n = len(A)
    C = np.zeros((n, n))
    for j in range(n):
        for k in range(n):
            for i in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

def multkji(A, B):
    n = len(A)
    C = np.zeros((n, n))
    for k in range(n):
        for j in range(n):
            for i in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

matrix_sizes = [20, 40, 60, 80, 100]

# элементы матрицы для multijk
execution_times = []
for size in matrix_sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)

    start_time = time.time()
    C1 = multijk(A, B)
    end_time = time.time()
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Отобразим график зависимости времени от порядка матриц для multijk
plt.plot(matrix_sizes, execution_times, label='multijk')

# элементы матрицы для multikj
execution_times = []
for size in matrix_sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)

    start_time = time.time()
    C2 = multikj(A, B)
    end_time = time.time()
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Отобразим график зависимости времени от порядка матриц для multikj
plt.plot(matrix_sizes, execution_times, label='multikj')

# элементы матрицы для multjik
execution_times = []
for size in matrix_sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)

    start_time = time.time()
    C3 = multjik(A, B)
    end_time = time.time()
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Отобразим график зависимости времени от порядка матриц для multjik
plt.plot(matrix_sizes, execution_times, label='multjik')

# элементы матрицы для multkij
execution_times = []
for size in matrix_sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)

    start_time = time.time()
    C4 = multkij(A, B)
    end_time = time.time()
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Отобразим график зависимости времени от порядка матриц для multkij
plt.plot(matrix_sizes, execution_times, label='multkij')

# элементы матрицы для multjki
execution_times = []
for size in matrix_sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)

    start_time = time.time()
    C5 = multjki(A, B)
    end_time = time.time()
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Отобразим график зависимости времени от порядка матриц для multjki
plt.plot(matrix_sizes, execution_times, label='multjki')

# элементы матрицы для multkji
execution_times = []
for size in matrix_sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)

    start_time = time.time()
    C6 = multkji(A, B)
    end_time = time.time()
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Отобразим график зависимости времени от порядка матриц для multkji
plt.plot(matrix_sizes, execution_times, label='multkji')

# элементы матрицы для @
execution_times = []
for size in matrix_sizes:
  A = np.random.rand(size, size)
  B = np.random.rand(size, size)
  start_time = time.time()
  C7 = A @ B
  end_time = time.time()
  execution_time = end_time - start_time
  execution_times.append(execution_time)
# Отобразим график зависимости времени от порядка матриц для @
plt.plot(matrix_sizes, execution_times, label='A@B')

plt.legend()
plt.xlabel('Размер матрицы')
plt.ylabel('Время выполнения (сек)')
plt.title('Зависимость времени от порядка матриц')

# Отображение графика
plt.show()
```

Вывод: Различное время выполнения каждой из написанных функций объясняется разными алгоритмическими подходами к умножению матриц.

То есть, каждая функция по разному выполняет итерации умножения, из-за чего как раз и идет разное время выполнения. А функция @ использует наиболее эффективные алгоритмы, из-за чего значитально быстрей остальных функций.
